5-1 리스트

```python
# 리스트 []

# 지하철 칸별로 10명, 20명, 30명
subway1 = 10
subway2 = 20
subway3 = 30

subway = [10, 20, 30] #subway라는 1개의 리스트안에 subway1, 2, 3을 가둬둔것
print(subway)

subway = ["유재석", "조세호", "박명수"]
print(subway)

#조세호씨가 몇 번째 칸에 타고 있는가? (~.ndex)
print(subway.index("조세호")) #0 1 2 순이라 1이 출력됌(조세호의 위치를 찾는 코드)

#하하씨가 다음 정류장에서 다음 칸에 탐 (~.append)
subway.append("하하")
print(subway)

# 정형돈씨를 유재석 / 조세호 사이에 태워봄 (~.insert)
subway.insert(1, "정형돈") # 태울 칸을 앞에 쓰고 , 한뒤에 이름 적을것
print(subway)

#지하철에 있는 사람을 한명씩 뒤에서 꺼냄 (~.pop)
print(subway.pop())
print(subway)

#print(subway.pop())
#print(subway)

#print(subway.pop())
#print(subway)

#같은 이름의 사람이 몇 명 있는지 확인 (~.count)
subway.append("유재석")
print(subway)
print(subway.count("유재석"))

#정렬도 가능 (~.sort)
num_list = [5, 2, 4, 3, 1]
num_list.sort() # sort는 정렬
print(num_list)

# 순서 뒤집기 가능 (~.reverse)
num_list.reverse() #reverse 뒤집는거
print(num_list)

# 모두 지우기 (~.clear)
num_list.clear()
print(num_list)

# 다양한 자료형 함께 사용
mix_list = ["조세호", 20, True]
#print(mix_list)
num_list = [5, 2, 4, 3, 1]

#리스트 확장 (~.extend)
num_list.extend(mix_list)
print(num_list)
```

5-4 세트

```python
# 집합 (set)
# ㄴ 중복 안됨, 순서 없음
my_set = {1,2,3,3,3} #사전에서는 key와 value를 썼는데 세트에서는 값만 나열하면 됌
print(my_set) # 중복이 안되서 1,2,3만 출력되는거

java = {"유재석", "김태호", "양세형"}
python = set(["유재석", "박명수"]) #리스트를 만들고나서 set으로 감쌈 #똑같이 정의할수있음 근데 위는 사전 아래는 집합임

# 교집합 (java와 python을 모두 할 수 있는 사람 and)
print(java & python)
print(java.intersection(python)) #intersection()은 교집합 (공통 요소) 을 구하는 메서드

# 합집합 (java 할수있거나 python 할수있는 개발자 or)
print(java | python)
print(java.union(python)) # union() 두 집합의 모든 요소를 합친 것

# 차집합 (java는 할줄 알지만 python은 할줄 모르는 개발자 difference)
print(java - python)
print(java.difference(python)) #difference() 한 집합에서 다른 집합에 없는 요소만

# python 할 줄 아는 사람이 늘어남
python.add("김태호") #add() 변수 추가
print(python)

# java를 까먹은 사람
java.remove("김태호") # remove() 변수 삭제
print(java)
```

5-5 자료구조의 변경

```python
# 자료구조의 변경

#커피숍
menu = {"커피", "우유", "주스"} #집합으로 일단 만듬
print(menu, type(menu)) # type을 쓰면 class가 나옴 집합이 나옴

menu = list(menu)
print(menu, type(menu)) # type을 리스트로 바꿔서 리스트가 나옴

menu = tuple(menu) #type을 tuple로 바꿔서 tuple이 나옴
print(menu, type(menu))

menu = set(menu) # 다시 복구
print(menu, type(menu))
```
